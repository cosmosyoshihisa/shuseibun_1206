```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
seiseki = pd.read_csv('../input/seiseki.csv')
```

```python
seiseki["dai1"] = 0
seiseki["dai2"] = 0
seiseki["dai3"] = 0
seiseki["dai4"] = 0
```

```python
seiseki["eigo"][1]
```

```python
for i in range(0,165,1):
    if seiseki["taiiku"][i]>50:
        if seiseki["eigo"][i]>50:
            seiseki["dai1"][i] = 1
        else:
            seiseki["dai2"][i] = 1
    else:
        if seiseki["eigo"][i]>50:
            seiseki["dai4"][i] = 1
        else:
            seiseki["dai3"][i] = 1
```
