```r
seiseki2 <- seiseki[,c(10,15,16,17,18)]
seiseki3 <- seiseki[,c(11,15,16,17,18)]
seiseki4 <- seiseki[,c(12,15,16,17,18)]
seiseki5 <- seiseki[,c(13,15,16,17,18)]
seiseki6 <- seiseki[,c(14,15,16,17,18)]
seiseki7 <- seiseki[,c(7,15,16,17,18)]
```

```r
pca2 <- prcomp(seiseki2,scale=TRUE)
pca3 <- prcomp(seiseki3,scale=TRUE)
pca4 <- prcomp(seiseki4,scale=TRUE)
pca5 <- prcomp(seiseki5,scale=TRUE)
pca6 <- prcomp(seiseki6,scale=TRUE)
pca7 <- prcomp(seiseki7,scale=TRUE)
```

```r
summary(pca2)
summary(pca3)
summary(pca4)
summary(pca5)
summary(pca6)
summary(pca7)
```
